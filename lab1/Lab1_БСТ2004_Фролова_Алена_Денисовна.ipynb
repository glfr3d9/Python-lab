## Функции:
### Сортировка выбором

```python
def Selection_sort(matrix):
    for row in range(len(matrix)):
        for col in range(len(matrix[0])-1):
            for colS in range(col+1, len(matrix[0])):
                if matrix[row][colS] < matrix[row][col]:
                    matrix[row][col], matrix[row][colS] = matrix[row][colS], matrix[row][col]
    return matrix
```

### Сортировка вставкой

```python
def insertion_sort(matrix):
    for row in range(len(matrix)):
        for col in range (1,len(matrix[0])):
            element = matrix[row][col]
            j = col-1
            while j >=0 and element < matrix[row][j]:
                matrix[row][j+1] = matrix[row][j]
                j -= 1
            matrix[row][j+1] = element
    return matrix
```

### Сортировка обменом

```python
def change_sort(matrix):
    for row in range(len(matrix)):
        for col in range(len(matrix[0])-1):
            for j in range(len(matrix[0])-col-1):
                if matrix[row][j] > matrix[row][j+1]:
                    matrix[row][j], matrix[row][j+1] = matrix[row][j+1], matrix[row][j]
    return matrix
```

### Сортировка Шелла

```python
def shell_sort(matrix):
    n = len(matrix[0])
    k = int(math.log2(n))
    interval = int(2**k -1)
    while interval > 0:
        for row in range(len(matrix)):
            for col in range(interval, n):
                element = matrix[row][col]
                j = col
                while j >= interval and matrix[row][j - interval] > element:
                    matrix[row][j] = matrix[row][j - interval]
                    j -= interval
                matrix[row][j] = element
        k -= 1
        interval = int(2**k -1)
    return matrix
```

### Турнирная сортировка

```python
def tournament_sort(matrix):
    el = [None] * 2 * (len(matrix) + len(matrix) % 2)
    index = len(el) - len(matrix) - len(matrix) % 2

    for i, v in enumerate(matrix):
        el[index + i] = (i, v)

    for j in range(len(matrix)):
        n = len(matrix)
        index = len(el) - len(matrix) - len(matrix) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1)
                if el[i] != None and el[i + 1] != None:
                    if el[i][1] < el[i + 1][1]:
                        el[i // 2] = el[i]
                    else:
                        el[i // 2] = el[i + 1]
                else:
                    el[i // 2] = el[i] if el[i] != None else el[i + 1]
            index -= n

        index, x = el[0]
        matrix[j] = x
        el[len(el) - len(matrix) - len(matrix) % 2 + index] = None
```

```python
def tournament_sort1(matrix):
    for i in range(len(matrix)):
        tournament_sort(matrix[i])
    return matrix
```

### Пирамидальная сортировка

```python
def heapify_sort(matrix, n, i, row):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2
    
    if l < n and matrix[row][i] < matrix[row][l]:
        largest = l
    if r < n and matrix[row][largest] < matrix[row][r]:
        largest = r
    
    if largest != i:
        matrix[row][i], matrix[row][largest] = matrix[row][largest], matrix[row][i]
        heapify_sort(matrix, n, largest, row)
```

### Быстрая сортировка

```python
def q_sort(matrix, low, high,row):
    pivot = matrix[row][(low + high) // 2]
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while matrix[row][i] < pivot:
            i += 1

        j -= 1
        while matrix[row][j] > pivot:
            j -= 1

        if i >= j:
            return j

        matrix[row][i], matrix[row][j] = matrix[row][j], matrix[row][i]
```

```python
def quick_sort(matrix):
    def quick_sort1(items, low, high,row):
        if low < high:
            split_index = q_sort(items, low, high,row)
            quick_sort1(items, low, split_index,row)
            quick_sort1(items, split_index + 1, high,row)

    for row in range(len(matrix)):
        quick_sort1(matrix, 0, len(matrix[row])-1,row)

    return matrix
```
